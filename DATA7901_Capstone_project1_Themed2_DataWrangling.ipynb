```python
import pandas as pd
import numpy as np
```

--add FlightNumber column and drop column id

```python
original_data = pd.read_csv("launch_sorted.csv")
original_data.insert(0, 'FlightNumber', range(1, len(original_data) + 1))
original_data = original_data.drop('id', axis=1)
original_data
```

```python
original_data.isnull().sum()
```

# Filter the dataframe to only include Falcon 9

--drop BoosterVersion = Falcon1

```python
original_data.drop(original_data.index[original_data['BoosterVersion'] == 'Falcon 1'], inplace = True)
data_falcon9= original_data
data_falcon9
```

```python
data_falcon9.loc[:,'FlightNumber'] = list(range(1, data_falcon9.shape[0]+1))
data_falcon9
```

```python
data_falcon9.isnull().sum()
```

--drop null value in column Orbit

```python
data_falcon9.dropna(subset=['Orbit'],inplace=True)
```

---Imputation the missing numerical value by the median of the
featureâ€™s class-specific values

```python
grouped = data_falcon9.groupby(['Orbit'])
class_specific_averages = grouped['PayloadMass'].transform('median')
data_falcon9['PayloadMass'] = data_falcon9['PayloadMass'].fillna(class_specific_averages)
data_falcon9
```

```python
data_falcon9.dropna(subset=['PayloadMass'],inplace=True)
```

```python
data_falcon9.drop(columns=['LandingPad'], inplace=True)
```

```python
data_falcon9
```

```python
data_falcon9.isnull().sum()
```

```python
data_falcon9.Outcome.value_counts()
```

```python
data_falcon9.dtypes
```

```python
landing_outcomes = data_falcon9.Outcome.value_counts()
```

```python
# Define the set of bad outcomes
fail_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]])

# Create the 'Class' column using apply and lambda
data_falcon9['Class'] =data_falcon9['Outcome'].apply(lambda x: 0 if x in fail_outcomes else 1)
```
